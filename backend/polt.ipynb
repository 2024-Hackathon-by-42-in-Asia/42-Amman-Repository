```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Download data from https://www.kaggle.com/datasets/debajyotipodder/co2-emission-by-vehicles
file_path = '../resources/data/CO2_emissions.csv'
df = pd.read_csv(file_path)
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Fuel Consumption City (L/100 km)', y='CO2 Emissions(g/km)', hue='Fuel Type')
sns.regplot(data=df, x='Fuel Consumption City (L/100 km)', y='CO2 Emissions(g/km)', scatter=False, color='red')
plt.title('Fuel Consumption in City vs CO2 Emissions by Fuel Type with Regression Line')
plt.xlabel('Fuel Consumption City (L/100 km)')
plt.ylabel('CO2 Emissions (g/km)')
plt.legend(title='Fuel Type')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Fuel Type', y='CO2 Emissions(g/km)', estimator='mean')
plt.title('Average CO2 Emissions per Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average CO2 Emissions (g/km)')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Fuel Consumption Comb (L/100 km)', y='CO2 Emissions(g/km)', hue='Vehicle Class')
sns.regplot(data=df, x='Fuel Consumption Comb (L/100 km)', y='CO2 Emissions(g/km)', scatter=False, color='red')
plt.title('Fuel Consumption Combined vs CO2 Emissions by Vehicle Class with Regression Line')
plt.xlabel('Fuel Consumption Comb (L/100 km)')
plt.ylabel('CO2 Emissions (g/km)')
plt.legend(title='Vehicle Class')
plt.show()
```
